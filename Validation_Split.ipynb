## Basic Preprocessing

```python
df = pd.read_csv("diamonds_whole.csv")
validation_split = pd.read_csv("diamonds_validation.csv")
training_split = pd.read_csv("diamonds_training.csv")
```

```python
#Replace values in cut
df = df.replace(to_replace ="Fair", value = 1)
df = df.replace(to_replace ="Good", value = 2)
df = df.replace(to_replace ="Very Good", value = 3)
df = df.replace(to_replace ="Premium", value = 4)
df = df.replace(to_replace ="Ideal", value = 5)
print(df["cut"].unique())

training_split = training_split.replace(to_replace ="Fair", value = 1)
training_split = training_split.replace(to_replace ="Good", value = 2)
training_split = training_split.replace(to_replace ="Very Good", value = 3)
training_split = training_split.replace(to_replace ="Premium", value = 4)
training_split = training_split.replace(to_replace ="Ideal", value = 5)
print(training_split["cut"].unique())

validation_split = validation_split.replace(to_replace ="Fair", value = 1)
validation_split = validation_split.replace(to_replace ="Good", value = 2)
validation_split = validation_split.replace(to_replace ="Very Good", value = 3)
validation_split = validation_split.replace(to_replace ="Premium", value = 4)
validation_split = validation_split.replace(to_replace ="Ideal", value = 5)
print(validation_split["cut"].unique())
```

```python
df = df.replace(to_replace ="J", value = 1)
df = df.replace(to_replace ="I", value = 2)
df = df.replace(to_replace ="H", value = 3)
df = df.replace(to_replace ="G", value = 4)
df = df.replace(to_replace ="F", value = 5)
df = df.replace(to_replace ="E", value = 6)
df = df.replace(to_replace ="D", value = 7)
print(df["color"].unique())

training_split = training_split.replace(to_replace ="J", value = 1)
training_split = training_split.replace(to_replace ="I", value = 2)
training_split = training_split.replace(to_replace ="H", value = 3)
training_split = training_split.replace(to_replace ="G", value = 4)
training_split = training_split.replace(to_replace ="F", value = 5)
training_split = training_split.replace(to_replace ="E", value = 6)
training_split = training_split.replace(to_replace ="D", value = 7)
print(training_split["color"].unique())

validation_split = validation_split.replace(to_replace ="J", value = 1)
validation_split = validation_split.replace(to_replace ="I", value = 2)
validation_split = validation_split.replace(to_replace ="H", value = 3)
validation_split = validation_split.replace(to_replace ="G", value = 4)
validation_split = validation_split.replace(to_replace ="F", value = 5)
validation_split = validation_split.replace(to_replace ="E", value = 6)
validation_split = validation_split.replace(to_replace ="D", value = 7)
print(validation_split["color"].unique())
```

```python
df = df.replace(to_replace ="I1", value = 1)
df = df.replace(to_replace ="SI2", value = 2)
df = df.replace(to_replace ="SI1", value = 3)
df = df.replace(to_replace ="VS2", value = 4)
df = df.replace(to_replace ="VS1", value = 5)
df = df.replace(to_replace ="VVS2", value = 6)
df = df.replace(to_replace ="VVS1", value = 7)
df = df.replace(to_replace ="IF", value = 8)
print(df["clarity"].unique())

training_split = training_split.replace(to_replace ="I1", value = 1)
training_split = training_split.replace(to_replace ="SI2", value = 2)
training_split = training_split.replace(to_replace ="SI1", value = 3)
training_split = training_split.replace(to_replace ="VS2", value = 4)
training_split = training_split.replace(to_replace ="VS1", value = 5)
training_split = training_split.replace(to_replace ="VVS2", value = 6)
training_split = training_split.replace(to_replace ="VVS1", value = 7)
training_split = training_split.replace(to_replace ="IF", value = 8)
print(training_split["clarity"].unique())

validation_split = validation_split.replace(to_replace ="I1", value = 1)
validation_split = validation_split.replace(to_replace ="SI2", value = 2)
validation_split = validation_split.replace(to_replace ="SI1", value = 3)
validation_split = validation_split.replace(to_replace ="VS2", value = 4)
validation_split = validation_split.replace(to_replace ="VS1", value = 5)
validation_split = validation_split.replace(to_replace ="VVS2", value = 6)
validation_split = validation_split.replace(to_replace ="VVS1", value = 7)
validation_split = validation_split.replace(to_replace ="IF", value = 8)
print(validation_split["clarity"].unique())
```

```python
df.to_csv("diamonds_whole_pp_1.csv", index=False)
training_split.to_csv("diamonds_tr_pp_1.csv", index=False)
validation_split.to_csv("diamonds_val_pp_1.csv", index=False)
```

```python
train = training_split.loc[:round(len(training_split)*0.8)]
test = training_split.loc[round(len(training_split)*0.8)+1:]
len(training_split), len(train), len(test)
```

```python
train.to_csv("diamonds_tr_80_pp_1.csv", index=False)
test.to_csv("diamonds_tr_20_pp_1.csv", index=False)
```
